# Python Installation and Setup Lecture

Welcome to your first Python programming lecture! Today, we'll cover how to install Python, set up your environment, and start writing your first lines of code.

## 1. Installing Python

### Step-by-Step Installation

1. **Download Python:**
   - Visit the [official Python website](https://www.python.org/downloads/).
   - Download the latest version for your operating system (Windows, macOS, or Linux).

2. **Install Python:**
   - Run the installer.
   - **Windows Users:** Ensure you check "Add Python to PATH" during installation.
   - Follow the installation prompts.

3. **Verify Installation:**
   - Open a terminal or command prompt.
   - Type `python --version` or `python3 --version` to verify the installation.

## 2. Setting Up Your Environment

### Choosing an Integrated Development Environment (IDE)

- **IDEs and Editors:**
  - **IDLE:** Comes installed with Python; great for beginners.
  - **VS Code:** A powerful, free editor with Python support.
  - **PyCharm:** A professional IDE with many features (Community Edition is free).

### Installing VS Code (Recommended)

1. **Download VS Code:**
   - Visit the [VS Code website](https://code.visualstudio.com/).
   - Download and install for your operating system.

2. **Install Python Extension:**
   - Open VS Code.
   - Go to the Extensions view (`Ctrl+Shift+X` or `Cmd+Shift+X` on macOS).
   - Search for "Python" and install the extension by Microsoft.

## 3. Running Python Code

### Using the Python Interpreter

- **Interactive Mode:**
  - Open a terminal or command prompt.
  - Type `python` or `python3` to start the interactive interpreter.
  - You can type and execute Python commands directly.

### Writing and Running Scripts

1. **Creating a Script:**
   - Open your IDE or text editor.
   - Create a new file with a `.py` extension, e.g., `hello.py`.
   - Write your Python code. Example:
     ```python
     print("Hello, World!")
     ```

2. **Running the Script:**
   - In the terminal, navigate to the directory containing your script.
   - Run `python hello.py` or `python3 hello.py`.

## 4. Writing Your First Program

### Example Program: Hello, World!

1. **Open Your Editor:**
   - Use IDLE, VS Code, or PyCharm.

2. **Write Code:**
   - Type the following in a new file:
     ```python
     print("Hello, World!")
     ```

3. **Save and Run:**
   - Save the file as `hello.py`.
   - Run the script through your terminal or the run feature in your IDE.
